# Exploratory Data Analysis(EDA)

**Exploratory Data Analysis (EDA)** is the process of analyzing and visualizing datasets to summarize their main characteristics, often using statistical graphics and other data visualization methods. EDA helps to:

- Understand the structure, patterns, and relationships within the data
- Detect anomalies, missing values, or outliers
- Formulate hypotheses and guide further analysis or modeling steps

Common EDA techniques include summary statistics, histograms, box plots, scatter plots, and correlation analysis.

# 1. Import libraries
